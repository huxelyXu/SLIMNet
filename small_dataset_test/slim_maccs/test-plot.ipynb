```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 假设数据已经定义
x = [4, 2, 1, 0, -1, -2, -3]
y = [np.mean(x00001), np.mean(x001), np.mean(x01), np.mean(x1), np.mean(x10), np.mean(x100), np.mean(x1000)]
yerr = [np.var(x00001), np.var(x001), np.var(x01), np.var(x1), np.var(x10), np.var(x100), np.var(x1000)]

# 设置 seaborn 风格，使配色更符合学术论文风格
sns.set(style="ticks", font_scale=1.2)

# 创建图形并设置大小
plt.figure(figsize=(5.5, 4), dpi=500)

# 绘制 errorbar
plt.errorbar(
    x,
    y,
    yerr=yerr,
    fmt="o",  # 使用圆形点作为标记
    ecolor="gray",  # 设置误差线颜色为灰色
    color="royalblue",  # 设置数据点颜色为深蓝色
    elinewidth=1.5,  # 设置误差线宽度
    capsize=5,  # 设置误差线顶端大小
)

# 绘制数据点连接线
plt.plot(x, y, c="royalblue")  

# 设置坐标轴标签和字体大小
# plt.xlabel("X Label", fontsize=18)  # 设置 x 轴标签
# plt.ylabel("Y Label", fontsize=18)  # 设置 y 轴标签
plt.xticks(fontsize=15)  # 设置 x 轴刻度字体大小
plt.yticks(fontsize=15)  # 设置 y 轴刻度字体大小

# 设置坐标轴线条粗细
ax = plt.gca()
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)

# 调整布局，使图形更美观
plt.tight_layout()

# 显示图形
plt.show()
```

```python
import numpy as np
def cal_error(a,b,c,d):
    
    tmp = np.array([a,b,c,d])
    mean = np.mean(tmp, axis = 0)
    error = np.sqrt(np.var(tmp, axis = 0))

    return mean, error
```
